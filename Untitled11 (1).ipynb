```python
import numpy as np
import matplotlib.pyplot as plt
#Data 
months = np.array(["jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])

confirmed_cases = np.array([3000, 4500, 5400, 6600, 9000, 5000, 6500, 5400, 6000, 7500, 4800, 8700])
recovered_cases  = np.array([600, 900, 1500, 2400, 3000, 2500, 2300, 3600, 1500, 2500, 2600, 3200])
death_cases      = np.array([200, 350, 450, 650, 780, 540, 560, 440, 740, 240, 870, 390])

active_cases = confirmed_cases - recovered_cases - death_cases

total_confirmed = np.sum(confirmed_cases)
average_cases = np.mean(confirmed_cases)
max_cases = np.max(confirmed_cases)

print("Total Confirmed Cases:", total_confirmed)
print("Average Daily Cases:", average_cases)
print("Maximum Cases in a Day:", max_cases)

#line Graph
plt.figure()
plt.plot(months, confirmed_cases, marker='o')
plt.xlabel("Days")
plt.ylabel("Confirmed Cases")
plt.title("Monthly Confirmed Cases in 2018")
plt.grid(True)
plt.show()

#Bar Chart
categories = ["Active", "Recovered", "Deaths"]
values = [active_cases[-1], recovered_cases[-1], death_cases[-1]]

plt.figure()
plt.bar(categories, values)
plt.xlabel("Case Type")
plt.ylabel("Number of Cases")
plt.title("COVID Case Distribution (In 2018)")
plt.show()

total_active = np.sum(active_cases)
total_recovered = np.sum(recovered_cases)
total_deaths = np.sum(death_cases)

plt.figure()
plt.pie(
    [total_active, total_recovered, total_deaths],
    labels=["Active", "Recovered", "Deaths"],
    autopct='%1.1f%%',
    explode = [0,0,0.1],
    colors = ["#c1efef","#3dccf5","#dd2929"],
    startangle=90,
)
plt.title("Overall COVID Case Distribution")
plt.show()
```
